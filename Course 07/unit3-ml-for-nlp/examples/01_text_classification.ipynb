## Code Example | Ù…Ø«Ø§Ù„ Ø§Ù„ÙƒÙˆØ¯

```python
"""
Unit 3 - Example 1: Text Classification with ML
Ø§Ù„ÙˆØ­Ø¯Ø© 3 - Ù…Ø«Ø§Ù„ 1: ØªØµÙ†ÙŠÙ Ø§Ù„Ù†Øµ Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… ØªØ¹Ù„Ù… Ø§Ù„Ø¢Ù„Ø©

This example demonstrates:
1. Feature extraction (TF-IDF)
2. Text classification
3. Model evaluation
"""

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

print("=" * 60)
print("Example 1: Text Classification with ML")
print("Ù…Ø«Ø§Ù„ 1: ØªØµÙ†ÙŠÙ Ø§Ù„Ù†Øµ Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… ØªØ¹Ù„Ù… Ø§Ù„Ø¢Ù„Ø©")
print("=" * 60)

# Sample data: Movie reviews
# Ø¨ÙŠØ§Ù†Ø§Øª Ù†Ù…ÙˆØ°Ø¬ÙŠØ©: Ù…Ø±Ø§Ø¬Ø¹Ø§Øª Ø§Ù„Ø£ÙÙ„Ø§Ù…
documents = [
    "This movie is amazing and wonderful",
    "I love this film, it's fantastic",
    "Terrible movie, very boring",
    "Worst film I've ever seen",
    "Great acting and story",
    "Poor quality, not recommended",
    "Excellent cinematography",
    "Boring and slow paced"
]

labels = [
    "positive", "positive", "negative", "negative",
    "positive", "negative", "positive", "negative"
]

print("\nSample Documents:")
print("Ù…Ø³ØªÙ†Ø¯Ø§Øª Ù†Ù…ÙˆØ°Ø¬ÙŠØ©:")
for i, (doc, label) in enumerate(zip(documents, labels), 1):
    print(f"  {i}. [{label}] {doc}")

# 1. Feature Extraction with TF-IDF
# Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ø§Ù„Ù…ÙŠØ²Ø§Øª Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… TF-IDF
print("\n" + "=" * 60)
print("1. Feature Extraction (TF-IDF)")
print("Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ø§Ù„Ù…ÙŠØ²Ø§Øª (TF-IDF)")
print("=" * 60)

vectorizer = TfidfVectorizer(max_features=10, stop_words='english')
X = vectorizer.fit_transform(documents)

print(f"\nFeature matrix shape: {X.shape}")
print(f"Features (words): {vectorizer.get_feature_names_out()}")

# 2. Train-Test Split
# ØªÙ‚Ø³ÙŠÙ… Ø§Ù„ØªØ¯Ø±ÙŠØ¨ ÙˆØ§Ù„Ø§Ø®ØªØ¨Ø§Ø±
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.25, random_state=42
)

print(f"\nTraining samples: {len(y_train)}")
print(f"Test samples: {len(y_test)}")

# 3. Train Classifier
# ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù…ØµÙ†Ù
print("\n" + "=" * 60)
print("2. Training Classifier")
print("ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù…ØµÙ†Ù")
print("=" * 60)

classifier = MultinomialNB()
classifier.fit(X_train, y_train)

print("âœ“ Classifier trained successfully")
print("âœ“ ØªÙ… ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù…ØµÙ†Ù Ø¨Ù†Ø¬Ø§Ø­")

# 4. Make Predictions
# Ø¹Ù…Ù„ Ø§Ù„ØªÙˆÙ‚Ø¹Ø§Øª
predictions = classifier.predict(X_test)

print("\n" + "=" * 60)
print("3. Predictions and Evaluation")
print("Ø§Ù„ØªÙˆÙ‚Ø¹Ø§Øª ÙˆØ§Ù„ØªÙ‚ÙŠÙŠÙ…")
print("=" * 60)

print("\nPredictions:")
print("Ø§Ù„ØªÙˆÙ‚Ø¹Ø§Øª:")
for i, (true_label, pred_label) in enumerate(zip(y_test, predictions), 1):
    status = "âœ“" if true_label == pred_label else "âœ—"
    print(f"  {i}. True: {true_label}, Predicted: {pred_label} {status}")

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print(f"\nAccuracy: {accuracy * 100:.1f}%")
print(f"Ø§Ù„Ø¯Ù‚Ø©: {accuracy * 100:.1f}%")

print("\n" + "=" * 60)
print("Example completed successfully!")
print("ØªÙ… Ø¥ÙƒÙ…Ø§Ù„ Ø§Ù„Ù…Ø«Ø§Ù„ Ø¨Ù†Ø¬Ø§Ø­!")
print("=" * 60)

```
